```python
from matplotlib import pyplot as plt
from datetime import datetime
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
# Read CSV into dataframe and preview
file_csv = "data/Merged.csv"
merged_df = pd.read_csv(file_csv, encoding="ISO-8859-1")
merged_df.head()
```

```python
merged_df = merged_df.dropna(subset=['Weekly_Sales'])
print(merged_df.count())
```

```python
#Label-encode the 'IsHoliday' column - stopping with label encoding as this column has only 2 unique values
label_encoder = LabelEncoder()
merged_df['newHoliday'] = label_encoder.fit_transform(merged_df['IsHoliday'])

#Use get_dummies to one-hot encode the 'Type' column - one-hot encoding this column as it has >2 unique values
encoded_df = pd.get_dummies(merged_df, columns=["Type"])
```

```python
C_mat = encoded_df.corr()
fig = plt.figure(figsize = (15,15))

sb.heatmap(C_mat, vmin =-1, vmax = 1, square=True, cmap='YlGnBu', color='blue', annot=True, fmt='.1f')
plt.show()
```
